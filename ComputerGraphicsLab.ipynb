```python
students = {}

def add_student(name, marks):
  students[name] = marks

def get_marks(name):
  return students.get(name, "Student not found")

def print_sorted_students():
  for name in sorted(students.keys()):
    print(name, " : ", students[name])

add_student("Alice", 90)
add_student("Bob", 85)
add_student("Charlie", 95)
print("Alice marks: ", get_marks("Alice"))
print("\nSorted dictionary:")
print_sorted_students()
```

##Question-6

```python
import numpy as np
import matplotlib.pyplot as plt

def bresenham_line(x0, y0, x1, y1):
  dx = abs(x1 - x0)
  dy = abs(y1 - y0)
  shiftX = 1 if x0 < x1 else -1
  shiftY = 1 if y0 < y1 else -1
  p = 2 * dy - dx if dx > dy else 2 * dx - dy
  x, y = x0, y0

  points = []

  while True:
    points.append((x, y))
    if x == x1 and y == y1:
      break
    if p < 0:
      if dx > dy:
        p += 2 * dy
        x += shiftX
      else:
        p += 2 * dx
        y += shiftY
    else:
      x += shiftX
      y += shiftY
      p += 2 * (dy - dx) if dx > dy else 2 * (dx - dy)

  return points

x1, y1 = map(int, input("Enter first point (x1 y1): ").split())
x2, y2 = map(int, input("Enter second point (x2 y2): ").split())

line = bresenham_line(x1, y1, x2, y2)

worksheet = np.zeros((100, 100))

for x, y in line:
  worksheet[y, x] = 1

plt.imshow(worksheet, cmap='gray', origin="lower")
plt.show()
```

##Question-1

```python
import numpy as np
import matplotlib.pyplot as plt

# Taking Inputs
print("Enter the line segment endpoints:")
x1, y1 = map(float, input("x1 y1: ").split())
x2, y2 = map(float, input("x2 y2: ").split())

print("\nEnter the polygon coordinates in anti-clockwise:")
num_vertices = int(input("Enter the number of vertices: "))
vertices = []

for i in range(num_vertices):
    x, y = map(float, input(f"Vertex {i + 1}: ").split())
    vertices.append((x, y))

print("")

# Cyrus-Beck Algorithm
p1 = np.array([x1, y1])
p2 = np.array([x2, y2])
polygon = np.array(vertices)
D = p2 - p1  # Direction of line
t_entry, t_exit = 0.0, 1.0

for i in range(num_vertices):
    p3 = polygon[i]
    p4 = polygon[(i + 1) % num_vertices]
    edge = p4 - p3
    normal = np.array([-edge[1], edge[0]])  # Outward normal
    normal = normal / np.linalg.norm(normal)  # Normalize

    numerator = np.dot(normal, p3 - p1)
    denominator = np.dot(normal, D)

    if denominator == 0:
        print(f"Line is parallel to edge {p3} -> {p4}")
        continue

    t = numerator / denominator
    print("t =", t, "for edge", p3, "to", p4)

    if denominator > 0:
        t_entry = max(t_entry, t)
    else:
        t_exit = min(t_exit, t)

print("\nt_entry =", t_entry)
print("t_exit =", t_exit)

# Plotting
plt.figure(figsize=(8, 8))
polygon_closed = np.vstack((polygon, polygon[0]))  # Close the polygon
plt.plot(polygon_closed[:, 0], polygon_closed[:, 1], 'r-', label='Polygon')
plt.plot([p1[0], p2[0]], [p1[1], p2[1]], 'b--', label='Original Line')

if t_entry <= t_exit:
    print("\nLine intersects polygon")
    clipped_start = p1 + t_entry * D
    clipped_end = p1 + t_exit * D
    print("Clipped start:", clipped_start)
    print("Clipped end:", clipped_end)
    plt.plot([clipped_start[0], clipped_end[0]], [clipped_start[1], clipped_end[1]], 'g-', linewidth=2, label='Clipped Line')
else:
    print("\nLine does not intersect polygon")

plt.legend()
plt.grid(True)
plt.title("Cyrus-Beck Line Clipping")
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def transformations(x, y, px, py, theta, tx, ty):
    point_vector = np.array([[x], [y], [1]])

    T1 = np.array([[1, 0, -px],
                   [0, 1, -py],
                   [0, 0, 1]])

    R = np.array([[np.cos(theta), -np.sin(theta), 0],
                  [np.sin(theta),  np.cos(theta), 0],
                  [0, 0, 1]])

    T2 = np.array([[1, 0, px + tx],
                   [0, 1, py + ty],
                   [0, 0, 1]])

    transformed_point = T2 @ R @ T1 @ point_vector
    return transformed_point[:2].flatten()

# Input
n = int(input("Enter the number of vertices of the polygon: "))
vertices = []

for i in range(n):
    x, y = map(float, input(f"Vertex {i + 1}: ").split())
    vertices.append((x, y))

theta_deg = float(input("Enter the angle theta in degrees: "))
theta = np.deg2rad(theta_deg)

# Apply transformation
rotated_vertices = [transformations(x, y, 0, 0, theta, 0, 0) for (x, y) in vertices]

x1, y1 = map(float, input("Enter the point about to rotate: ").split())
theta1 = float(input("Enter the degree of rotation about the above point: "))
theta1 = np.deg2rad(theta1)

rotated_vertices1 = [transformations(x, y, x1, y1, theta1, 0, 0) for (x, y) in vertices]

# Close the polygon by repeating the first point
vertices.append(vertices[0])
rotated_vertices.append(rotated_vertices[0])
rotated_vertices1.append(rotated_vertices1[0])

# Split x and y for plotting
original_x, original_y = zip(*vertices)
rotated_x, rotated_y = zip(*rotated_vertices)
rotated_x1, rotated_y1 = zip(*rotated_vertices1)

# Plot
plt.figure(figsize=(6, 6))
plt.plot(original_x, original_y, 'bo-', label='Original Polygon')
plt.plot(rotated_x, rotated_y, 'ro--', label='Rotated Polygon')
plt.plot(rotated_x1, rotated_y1, 'go--', label='Rotated Polygon about a point')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Original and Rotated Polygon")
plt.legend()
plt.grid(True)
plt.show()
```

##Question-1

```python
import numpy as np
import matplotlib.pyplot as plt

# Input
x, y = map(float, input("Enter the point (x y): ").split())
theta = float(input("Enter the angle in degrees: "))
theta = np.deg2rad(theta)

# Rotation matrix and rotated point
rotated_matrix = [[np.cos(theta), -np.sin(theta)],
                  [np.sin(theta),  np.cos(theta)]]
sheared_vector = np.dot(rotated_matrix, np.array([x, y]))

# Plotting
plt.figure(figsize=(6, 6))
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)

# Original and rotated points
plt.plot(x, y, 'bo', label='Original Point')
plt.plot(sheared_vector[0], sheared_vector[1], 'ro', label='Rotated Point')

# Arrows from origin
plt.plot([0, x], [0, y], 'b--')
plt.plot([0, sheared_vector[0]], [0, sheared_vector[1]], 'r--')

# Text labels on the points
plt.text(x + 0.1, y + 0.1, f'({x:.2f}, {y:.2f})', color='blue')
plt.text(sheared_vector[0] + 0.1, sheared_vector[1] + 0.1,
         f'({sheared_vector[0]:.2f}, {sheared_vector[1]:.2f})', color='red')

# Plot settings
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Rotation of a Point")
plt.legend()
plt.grid(True)
plt.show()
```
